## Filtering

1. Read the test image
2. Define the identity kernel, using a 3Ã—3 NumPy array
3. Use the filter2D() function in OpenCV to perform the linear filtering operation
4. Display the original and filtered images, using imshow()
5. Save the filtered image to disk, using imwrite()

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

**filtered_img = filter2D(src, ddepth, kernel)**

* src : the source image
* ddepth : Depth of the resulting image. 
    * -1 indicates that the final image will also have the same depth as the source image
* kernel : kernel for convolution

```python
image = cv2.imread('Images/wood.jpg',cv2.IMREAD_GRAYSCALE)
"""
Apply identity kernel
"""
kernel1 = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])

# filter2D() function can be used to apply kernel to an image.
# Where ddepth is the desired depth of final image. ddepth is -1 if...
# ... depth is same as original or source image.

identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)

# Display the image
# plt.imshow(image,cmap='gray')
plt.imshow(identity,cmap='gray')
```

```python
"""
Apply sharpening using kernel
"""
sharping = np.array([[0, -1,  0],
                   [-1,  5, -1],
                    [0, -1,  0]])
sharp_img = cv2.filter2D(src=image, ddepth=-1, kernel=sharping)

# Display the image
# plt.imshow(image,cmap='gray')
plt.imshow(sharp_img,cmap='gray')
```

**img_blur = cv2.blur(src,ksize)**

* src : source image
* ksize : kernel size 

```python
# Apply blur using `blur()` function
img_blur = cv2.blur(src=image, ksize=(3,3)) # Using the blur function to blur an image where ksize is the kernel size

# Display the image
plt.imshow(img_blur,cmap='gray')
```

**gaussian_blur = GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType]]])**


* src : Source image that you want to filter
* ksize : Size of the Gaussian kernel
* sigmaX and sigmaY : Gaussian kernel standard deviations, in the X (horizontal) and Y (vertical) direction. 

    * The default setting of sigmaY is zero. 
    * If you simply  set sigmaX to zero, then the standard deviations are computed from the kernel size (width and height respectively). Y
    * Size of each argument to positive values greater than zero.

```python
# sigmaX is Gaussian Kernel standard deviation 
# ksize is kernel size
gaussian_blur = cv2.GaussianBlur(src=image, ksize=(5,5), sigmaX=0, sigmaY=0)


# Display the image
plt.imshow(gaussian_blur,cmap='gray')
```

**median = medianBlur(src, ksize)**

* src : Source image.
* ksize : Kernel size, which must be an odd, positive integer.

```python
# medianBlur() is used to apply Median blur to image
# ksize is the kernel size
median = cv2.medianBlur(src=image, ksize=5)

# Display the image
plt.imshow(median,cmap='gray')
```

## Bilateral Bluring Filter = Sharping + Bluring 

**bilateral_filter = bilateralFilter(src, d, sigmaColor, sigmaSpace)**


* src : Source Image
* d : Diameter of the pixel neighborhood used for filtering
* sigmaColor : Standard deviation of the (1D) color-intensity distribution  
* sigmaSpace : Standard deviation of (2D) spatial distribution respectively

The sigmaSpace parameter defines the spatial extent of the kernel, in both the x and y directions (just like the Gaussian blur filter previously described).

The sigmaColor parameter defines the one-dimensional Gaussian distribution, which specifies the degree to which differences in pixel intensity can be tolerated.


```python
#Bilateral filter
bilateral_filter = cv2.bilateralFilter(src=image, d=9, sigmaColor=75, sigmaSpace=75)

# Display the image
plt.imshow(bilateral_filter,cmap='gray')
```
